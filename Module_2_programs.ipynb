Program to get the maximum and minimum value in a dictionary

```python
abc = {'a':6,'b':8,'c':34,'d':14}
z=abc['a']
for i in abc.values():
  if i>z:
    z=i

print(z)

x=abc['b']
for j in abc.values():
  if j<x:
    x=j

print(x)
```
