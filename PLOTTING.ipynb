```python
y = c.array([66, 27, 35, 25])
mylabels = ["a", "B", "C", "D"]
myexplode = [0.2, 0.1, 0.5, 0.7]

ns.pie(y, labels = mylabels, explode = myexplode)
ns.show() 
```
